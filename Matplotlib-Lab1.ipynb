```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'year': [2010, 2011, 2012, 2010, 2011, 2012, 2010, 2011, 2012],
        'team': ['FCBarcelona', 'FCBarcelona', 'FCBarcelona', 'RMadrid', 'RMadrid', 'RMadrid', 'ValenciaCF',
                 'ValenciaCF', 'ValenciaCF'],
        'wins':   [30, 28, 32, 29, 32, 26, 21, 17, 19],
        'draws':  [6, 7, 4, 5, 4, 7, 8, 10, 8],
        'losses': [2, 3, 2, 4, 2, 5, 9, 11, 11]}

football = pd.DataFrame(data, columns=['team','year','wins', 'draws', 'losses'])

football   
```

```python
plt.hist(football['losses'], facecolor = 'b')
plt.show()
```

```python




plt.bar(football['team'],football[ 'wins'])
plt.xlabel('team',fontsize = 20)
plt.ylabel('wins',fontsize = 20)
```

```python
football.groupby('team').sum().sort_values(by='wins',ascending = False)['wins'][0:5]
football.groupby('team').sum().sort_values(by='draws',ascending = False)['draws'][0:5]
```

```python
team=['FCBarcelona','RMadrid','ValenciaCF']
totalwins=[90,87,57]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(totalwins,labels=team, autopct='%1.1f%%', startangle=90, shadow=True,
        radius=1, explode = [0.3,0,0])


plt.show()
```

```python
plt.scatter(football['team'],football['wins'], marker = '*', s = 50,alpha = 0.4, label = 'team') 
plt.scatter(football['team'],football['draws'], marker = '+', s = 50,alpha = 0.4, label ='team')


plt.show()
```

```python
#Q4: Add the following to your chart:
plt.scatter(football['team'],football['losses'], alpha = 0.3) 
x='red'
plt.title('team of losses',color =x,loc='left' )
plt.xlabel('team',fontsize = 16)
plt.ylabel('losses',fontsize = 16)


plt.show()
```
